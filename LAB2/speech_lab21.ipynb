```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
filename="cherry.wav"
ipd.Audio(filename)
```

**A1. Find the first derivative of your speech signal with finite difference method. Listen to the first derivative signal and the original speech signal.**.


```python
# Calculate the finite difference derivative
def firstderivative1(signal, fs):
    derivative = np.diff(signal) * fs
    return derivative

# Load the speech signal
audio_data1, fs1 = librosa.load(filename)

# Calculate the first derivative using finite difference
derivative_signal1 = firstderivative1(audio_data1, fs1)

# Plot the original and derivative signals
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.plot(audio_data1)
plt.title('Original Speech Signal')

plt.subplot(1, 2, 2)
plt.plot(derivative_signal1)
plt.title('First Derivative Signal')
plt.tight_layout()
plt.show()

# Listen to the original and derivative signals
print("Original Speech Signal:")
display(Audio(data=audio_data1, rate=fs1))

print("Derivative Signal:")
display(Audio(data=derivative_signal1, rate=fs1))
```

 **A2. Detect the points of zero crossing in the first derivative signal.
Compare the average length between two consecutive zero crossings for speech and silence regions. Observe the pattern.**


```python
# Calculate the finite difference derivative
def firstderivative2(signal, fs):
  derivative = np.diff(signal) * fs
  return derivative
```

```python
# Find indices where the signal crosses zero
def zero_crossings2(signal):
    zero_crossings = np.where(np.diff(np.sign(signal)))[0]
    return zero_crossings
```

```python
# Load the speech signal
audio_data2, fs2 = librosa.load(filename)
# Calculate the first derivative using finite difference
derivative_signal2 = firstderivative2(audio_data2, fs2)
# Find zero crossings in the derivative signal
zero_crossings_indices2 = zero_crossings2(derivative_signal2)
```

```python
# Plot the original speech signal, derivative signal, and zero crossings
plt.figure(figsize=(14, 8))

# Original speech signal
plt.subplot(3, 1, 1)
plt.plot(audio_data2)
plt.title('Original Speech Signal')

# First derivative signal
plt.subplot(3, 1, 2)
plt.plot(derivative_signal2)
plt.title('First Derivative Signal')

# Mark zero crossings on the derivative signal
plt.subplot(3, 1, 3)
plt.plot(derivative_signal2)
plt.scatter(zero_crossings_indices2, derivative_signal2[zero_crossings_indices2], c='r', marker='x', label='Zero Crossing')
plt.title('Zero Crossings in Derivative Signal')

plt.tight_layout()
plt.show()
```

```python
print(zero_crossings_indices2)
```

```python
# Calculate the average length between two consecutive zero crossings for speech and silence
speech_intervals = zero_crossings_indices2[::2]
# Assuming speech occurs at odd indices
silence_intervals = zero_crossings_indices2[1::2]
# Assuming silence occurs at even indices
average_length_speech = np.mean(np.diff(speech_intervals))
average_length_silence = np.mean(np.diff(silence_intervals))
print("Average Length between Consecutive Zero Crossings (Speech):", average_length_speech)
print("Average Length between Consecutive Zero Crossings (Silence):", average_length_silence)
```

```python
# Calculate the average length between consecutive zero crossings
average_length = np.mean(np.diff(zero_crossings_indices2))
print("Average Length between Consecutive Zero Crossings:", average_length)
```
