```python
import librosa
import librosa.display
import IPython.display as ipd
```

```python
filename="cherry.wav"
ipd.Audio(filename)
```

A1. Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal. You may use the below code from librosa as a reference.

```python
y, sr = librosa.load('cherry.wav')
```

```python
librosa.display.waveshow(y)
```

A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.

```python
import numpy as np
```

```python
length=len(y)
magnitude = (np.min(y), np.max(y))
print(f"Signal Length: {length} ")
print(f"Magnitude range: {magnitude}")
```

```python
duration = librosa.get_duration(y=y, sr=sr)
print(f"Time range: 0 to {duration} seconds")
```

A3. Take a small segment of the signal and play it.


```python
from IPython.display import Audio
```

```python
start =1.5
end= 3.2
segment_1 = y[int(start* sr):int(end* sr)]
print("Segment1 from 1.5 to 3.2:")
Audio(segment_1, rate=sr)
```

```python
librosa.display.waveshow(segment_1)
```

A4. Play around with your recorded speech signal for various segments. Understand the nature of the signal. Also observe with abruptly segmented speech, how the perception of that speech is affected.
